```python
app_code = """
from flask import Flask, jsonify, render_template
from flask_cors import CORS
import json

app = Flask(__name__)
CORS(app)

def load(file):
    with open(f"data/{file}") as f:
        return json.load(f)

@app.route("/")
def home():
    return render_template("index.html")

@app.route("/states")
def states():
    return jsonify(load("states.json"))

@app.route("/dishes")
def dishes():
    return jsonify(load("dishes.json"))

@app.route("/featured")
def featured():
    return jsonify(load("featured.json"))

if __name__ == "__main__":
    app.run()
"""

open("flavours-ne/app.py","w").write(app_code)
```
